```python
import pandas as pd
data = pd.read_csv("binary.csv")
data.head()
```

```python
import numpy as np
# Importing matplotlib
import matplotlib.pyplot as plt

# Function to help us plot
def plot_points(data):
    X = np.array(data[["gre","gpa"]])
    y = np.array(data["admit"])
    admitted = X[np.argwhere(y==1)]
    rejected = X[np.argwhere(y==0)]
    plt.scatter([s[0][0] for s in rejected], [s[0][1] for s in rejected], s = 25, color = 'red', edgecolor = 'k')
    plt.scatter([s[0][0] for s in admitted], [s[0][1] for s in admitted], s = 25, color = 'cyan', edgecolor = 'k')
    plt.xlabel('Test (GRE)')
    plt.ylabel('Grades (GPA)')
    
# Plotting the points
plot_points(data)
plt.show()
```

```python
data["gre"] /= 800.
data["gpa"] /= 4.
data = pd.concat([data, pd.get_dummies(data["admit"], prefix="admit")], axis=1)
data= data.drop("admit", axis=1)
data = pd.concat([data, pd.get_dummies(data["rank"], prefix="rank")], axis=1)
data= data.drop("rank", axis=1)
data.head()
```

```python
# import numpy as np
# import keras
# from keras.utils import np_utils
# X = np.array(data)[:, 1:]
# y = np_utils.to_categorical(np.array(data["admit"]))
# print(y)

X_header = [col for col in data if not col.startswith("admit")]
y_header = [col for col in data if col.startswith("admit")]
X = data[X_header]
y = data[y_header]
print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2018)
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Dropout

model = Sequential()
model.add(Dense(128, activation="relu", input_shape=(6,)))
model.add(Dropout(0.2))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.1))
model.add(Dense(2, activation="softmax"))

model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 200, batch_size = 100, verbose = 1)
```

```python
score = model.evaluate(X_test, y_test)
print("\nAccuracy: ", score)
print("\nPredictions:")
print(model.predict_proba(X_test))
```
